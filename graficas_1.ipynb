## Contar cafeteros y no cafeteros

```python
df2 = df.groupby(['coffee']).count()
df2.head()
```

```python
df2.columns
```

```python
df2.reset_index(inplace = True)
```

```python
df2.head()
```

```python
tamaño = len(df2)
tamaño
```

```python
datos = "[ "
datos
```

```python
i = 1
for index, row in df2.iterrows():
  coffee = row['coffee']
  value = row['tip']
  if i < tamaño:
    datos += '{ "coffee": "'+coffee+'", "value": '+str(value)+'},'
    i = i + 1
  else:
    datos += '{ "coffee": "'+coffee+'", "value": '+str(value)+'} ]'
```
